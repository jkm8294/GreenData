# 01 — Data Exploration

Explore the raw government data collected from EIA, NREL, NOAA, and USGS.
Verify data quality, check for nulls, and understand feature distributions.

```python
import sys
sys.path.insert(0, '../..')

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='darkgrid')
pd.set_option('display.max_columns', None)
```

```python
# Load master features
features = pd.read_parquet('../../data/processed/features.parquet')
print(f'Shape: {features.shape}')
features.head(10)
```

```python
# Data completeness
null_pct = (features.isnull().sum() / len(features) * 100).round(1)
print('Null percentage per column:')
print(null_pct.to_string())
```

```python
# Feature distributions
numeric = features.select_dtypes(include='number')
fig, axes = plt.subplots(3, 4, figsize=(16, 10))
for i, col in enumerate(numeric.columns[:12]):
    ax = axes[i // 4, i % 4]
    numeric[col].dropna().hist(bins=20, ax=ax, color='#22c55e', alpha=0.7)
    ax.set_title(col, fontsize=9)
    ax.tick_params(labelsize=7)
plt.suptitle('Feature Distributions', fontsize=14)
plt.tight_layout()
plt.show()
```
